```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_validate, train_test_split
from sklearn.naive_bayes import GaussianNB

import sys
print(sys.version)

golf_file = "golf.csv"

# Open the file for reading and read in data
golf_file_handler = open(golf_file, "r")
golf_data = pd.read_csv(golf_file_handler, sep=";")
golf_file_handler.close()

golf_data.head(100)
```

```python
# Task 1: replace the categorical data with numerical data,
# so that it can be used in sklearn.naive_bayes.
d = {'sunny': 1, 'overcast': 2, 'rainy': 3}
golf_data.Outlook = [d[item] for item in golf_data.Outlook.astype(str)]
golf_data.head(10)
```

```python
# Task 2: Split the datatset into 70% train and 30% test data.
top = golf_data[:int(0.3*len(golf_data))]
bot = golf_data[int(0.3*len(golf_data)):]
print(top)
print(bot)
```

```python
# Lets initialise Gaussian Naive Bayes.
# https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.GaussianNB.html 
naive_b = GaussianNB()

# Task 3: Train the naive bayes model with the parameters and the 
# classification to learn from using the fit method.
naive_b.fit(bot[['Outlook','Temperature','Humidity','Windy']], bot[['Play']])
```

```python
# Task 4: build a data frame to compare the expected and the predicted values 
# from the test dataframe you created earlier.

# Train the naive bayes model with the parameters and the classification to learn from
naive_b.fit(bot[['Outlook','Temperature','Humidity','Windy']], bot[['Play']])

# build a dataframe to show the expected vs predicted values
prediction_array=naive_b.predict(top[['Outlook','Temperature','Humidity','Windy']])
prediction_dataframe=pd.DataFrame(data=prediction_array, columns=["prediction"])
prediction_dataframe['result']=top['Play']

print (prediction_dataframe)
```

```python
# Task 5: Use the score function and output the prediction accuracy
# for the train data set and the test data set.

print ("Naive Bayes Accuracy:", naive_b.score(top[['Outlook','Temperature','Humidity','Windy']], top[['Play']]))
```
